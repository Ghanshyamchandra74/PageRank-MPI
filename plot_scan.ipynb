```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scan1 = pd.read_csv('data/Exec_time_MPI_Scan1.csv')
scan2 = pd.read_csv('data/Exec_time_MPI_Scan2.csv')
scan4 = pd.read_csv('data/Exec_time_MPI_Scan4.csv')
scan8 = pd.read_csv('data/Exec_time_MPI_Scan8.csv')
scan16 = pd.read_csv('data/Exec_time_MPI_Scan16.csv')
scan32 = pd.read_csv('data/Exec_time_MPI_Scan32.csv')
```

```python
T_scan_1 = scan1["t"]
T_scan_2 = scan2["t"]
T_scan_4 = scan4["t"]
T_scan_8 = scan8["t"]
T_scan_16 = scan16["t"]
T_scan_32 = scan32["t"]
```

# For MPI_Own

```python
Time_scan_10000 = [];
Time_scan_20000 = [];
Time_scan_30000 = [];
Time_scan_40000 = [];
Time_scan_50000 = [];
```

```python
Speed_scan_10000 = [];
Speed_scan_20000 = [];
Speed_scan_30000 = [];
Speed_scan_40000 = [];
Speed_scan_50000 = [];
```

```python
Time_scan_10000.append(T_scan_1[0]);
Time_scan_10000.append(T_scan_2[0]);
Time_scan_10000.append(T_scan_4[0]);
Time_scan_10000.append(T_scan_8[0]);
Time_scan_10000.append(T_scan_16[0]);
Time_scan_10000.append(T_scan_32[0]);

```

```python
Time_scan_20000.append(T_scan_1[1]);
Time_scan_20000.append(T_scan_2[1]);
Time_scan_20000.append(T_scan_4[1]);
Time_scan_20000.append(T_scan_8[1]);
Time_scan_20000.append(T_scan_16[1]);
Time_scan_20000.append(T_scan_32[1]);
```

```python
Time_scan_30000.append(T_scan_1[2]);
Time_scan_30000.append(T_scan_2[2]);
Time_scan_30000.append(T_scan_4[2]);
Time_scan_30000.append(T_scan_8[2]);
Time_scan_30000.append(T_scan_16[2]);
Time_scan_30000.append(T_scan_32[2]);
```

```python
Time_scan_40000.append(T_scan_1[3]);
Time_scan_40000.append(T_scan_2[3]);
Time_scan_40000.append(T_scan_4[3]);
Time_scan_40000.append(T_scan_8[3]);
Time_scan_40000.append(T_scan_16[3]);
Time_scan_40000.append(T_scan_32[3]);
```

```python
Time_scan_50000.append(T_scan_1[4]);
Time_scan_50000.append(T_scan_2[4]);
Time_scan_50000.append(T_scan_4[4]);
Time_scan_50000.append(T_scan_8[4]);
Time_scan_50000.append(T_scan_16[4]);
Time_scan_50000.append(T_scan_32[4]);
```

```python
speed_up_scan_10000 = Time_scan_10000[0]/Time_scan_10000
speed_up_scan_20000 = Time_scan_20000[0]/Time_scan_20000
speed_up_scan_30000 = Time_scan_30000[0]/Time_scan_30000
speed_up_scan_40000 = Time_scan_40000[0]/Time_scan_40000
speed_up_scan_50000 = Time_scan_50000[0]/Time_scan_50000
```

```python
num_procs = [1,2,4,8,16,32]
```

```python
plt.plot(num_procs,Time_scan_10000,label='Array size: 10000')
plt.bar(num_procs,Time_scan_10000)
plt.plot(num_procs,Time_scan_20000,label='Array size: 20000')
plt.bar(num_procs,Time_scan_20000)
plt.plot(num_procs,Time_scan_30000,label='Array size: 30000')
plt.bar(num_procs,Time_scan_30000)
plt.plot(num_procs,Time_scan_40000,label='Array size: 40000')
plt.bar(num_procs,Time_scan_40000)
plt.plot(num_procs,Time_scan_50000,label='Array size: 50000')
plt.bar(num_procs,Time_scan_50000)
plt.grid()
lgd = plt.legend(loc='upper right',bbox_to_anchor=(1.5, 1.02))
plt.xticks(num_procs)
plt.xlabel('Threads')
plt.ylabel('Execution Time(s)')
plt.title('PageRank')
plt.savefig('mpi_time_page_rank.eps', format='eps', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
#x = [2,4,8,18,32]
plt.plot(num_procs,speed_up_scan_10000,label='Array size: 10000')
plt.bar(num_procs,speed_up_scan_10000)
plt.plot(num_procs,speed_up_scan_20000,label='Array size: 20000')
plt.bar(num_procs,speed_up_scan_20000)
plt.plot(num_procs,speed_up_scan_30000,label='Array size: 30000')
plt.bar(num_procs,speed_up_scan_30000)
plt.plot(num_procs,speed_up_scan_40000,label='Array size: 40000')
plt.bar(num_procs,speed_up_scan_40000)
plt.plot(num_procs,speed_up_scan_50000,label='Array size: 50000')
plt.bar(num_procs,speed_up_scan_50000)
plt.grid()
lgd = plt.legend(loc='upper right',bbox_to_anchor=(1.4, 1.02))
plt.xticks(num_procs)
plt.xlabel('Threads')
plt.ylabel('Speed Up( X Sequential)')
plt.title('PageRank')
plt.savefig('mpi_speed_up_page_rank.eps', format='eps', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
