```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('data/Time_Exec_1_MPI.csv')
df2 = pd.read_csv('data/Time_Exec_2_MPI.csv')
df4 = pd.read_csv('data/Time_Exec_4_MPI.csv')
df8 = pd.read_csv('data/Time_Exec_8_MPI.csv')
df16 = pd.read_csv('data/Time_Exec_16_MPI.csv')
df32 = pd.read_csv('data/Time_Exec_32_MPI.csv')
```

```python
T_1 = df1["t"]
T_2 = df2["t"]
T_4 = df4["t"]
T_8 = df8["t"]
T_16 = df16["t"]
T_32 = df32["t"] 
```

```python
speed_up = [];
```

```python
speed_up.append(T_1[0])
speed_up.append(T_2[0])
speed_up.append(T_4[0])
speed_up.append(T_8[0])
speed_up.append(T_16[0])
speed_up.append(T_32[0])
```

```python
num_procs = [1,2,4,8,16,32]
```

```python
plt.plot(num_procs,speed_up,label='MPI Execution Time(s)')
plt.bar(num_procs,speed_up)
plt.grid()
lgd = plt.legend(loc='upper right',bbox_to_anchor=(1.5, 1.02))
plt.xticks(num_procs)
plt.xlabel('Threads')
plt.ylabel('Execution Time(s)')
plt.title('PageRank')
plt.savefig('mpi_time_page_rank.eps', format='eps', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
speed_up = speed_up[0]/speed_up
```
